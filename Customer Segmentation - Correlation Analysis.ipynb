# Customer Segmentation  - Correlation Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\arnol\Desktop\Customer_Segementation_End-to-End_Analysis\correlation.csv")
df
```

```python
# Selecting the columns with the numeric data types.

numeric_columns = df.select_dtypes(include=['int', 'float'])

print(numeric_columns)
```

```python
df2 = numeric_columns
df2
```

## Pearson Correlation coefficient

```python
correlation_matrix = df2.corr()
correlation_matrix
```

```python
# Creating the correlation matrix
correlation_matrix = df2.corr()

# Creating a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Pearson Correlation Coefficients')
plt.show()
```

```python
coverage_correlations = df2.corrwith(df2['Coverage Amount'])
plt.figure(figsize=(8, 6))
coverage_correlations.plot(kind='bar', color='skyblue')
plt.title('Correlation with Coverage Amount')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
premium_correlations = df2.corrwith(df2['Premium Amount'])
plt.figure(figsize=(8, 6))
premium_correlations.plot(kind='bar', color='skyblue')
plt.title('Correlation with Premium Amount')
plt.xlabel('Variables')
plt.ylabel('Correlation Coefficient')
plt.show()
```
